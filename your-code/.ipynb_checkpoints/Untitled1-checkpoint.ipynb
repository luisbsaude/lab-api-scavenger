```python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
from requests.auth import HTTPBasicAuth
from getpass import getpass
```

```python
url = 'https://api.github.com/repos/ironhack-datalabs/datalis0819/commits'
response=requests.get(url,auth = HTTPBasicAuth('luisbsaude', "0ac30be3a3a182c28425088d0e49f7839272fa9d"))
```

```python
df_commit=pd.DataFrame(response.json())
```

```python
norm_commit=json_normalize(response.json())
ds_commit_norm=pd.DataFrame(norm_commit)
ds_commit_norm
```

```python
df_count=ds_commit_norm.groupby("commit.author.name").count()[["sha"]]
df_count=df_count.reset_index()
```

```python
df_count.columns=["Name", "# Commits"]
df_count
```
